```python
import tensorflow as tf # modelos e ferramentas para ML E DL
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
import numpy as np # Ferramentes para tratar valores númericos e listas
print(tf.version.VERSION)
```

```python
import numpy as np

xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=np.float64)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=np.float64)
```

```python
print(xs)
print(ys)
```

```python
model = Sequential([Dense(units=1, input_shape=[1])])
```

```python
model.summary()
```

```python
model.compile(optimizer='sgd', loss='mse')
```

```python
teste = np.array([-1.0], dtype=np.float64)
saida = model.predict(teste)
print(saida)
```

```python
history = model.fit(x=xs,y=ys, epochs=500)
```
